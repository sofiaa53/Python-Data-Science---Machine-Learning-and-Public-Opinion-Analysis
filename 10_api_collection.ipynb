```python
import requests
import json
import matplotlib.pyplot as plt
import pandas as pd
```

```python
response = requests.get("https://api.covid19api.com/country/indonesia/status/confirmed/live?from=2021-12-15T00:00:00Z&to=2022-01-15T00:00:00Z")
print(response.status_code)
print(response.text)
```

```python
text = json.dumps(response.json(), sort_keys=True, indent=4)
print(text)
```

```python
data = response.json()

date = [i['Date'] for i in data[:31]]
cases =[i['Cases'] for i in data[:31]]

df = pd.DataFrame({'Dates':date, 'Cases':cases})
df['Dates'] = [pd.to_datetime(i) for i in df['Dates']]

print(df.sort_values(by='Dates'))
```

```python
plt.plot(date, cases)
plt.title('Grafik Kasus Covid Terkonfirmasi 15 Desember 2021 - 15 Januari 2022')
plt.xlabel('Dates')
plt.ylabel('Cases')
```
