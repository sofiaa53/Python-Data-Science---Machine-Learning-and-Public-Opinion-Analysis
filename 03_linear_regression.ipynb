```python
#import semua library dan package yang dibutuhkan disini
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split as tts
from sklearn.linear_model import LinearRegression as lr
from sklearn import metrics
```

```python
#tampilkan 5 data teratas
df_bensin = pd.read_csv("assets/bensin.csv")
display(df_bensin.head())
```

```python
#tampilkan informasi statistik deskriptif dari data
df_bensin.describe()
```

```python
liter = df_bensin[['Liter']]
jarak = df_bensin[['Kilometer']]

# x_train = menampung data source yang akan dilatih
# x_test = menampung data target yang akan dilatih
# y_train = menampung data source untuk testing
# y_test = menampung data target untuk testing
# test_size 0.2 berarti dataset digunakan 80% untuk data training dan 20% untuk data test (default nilai test_size=None berarti 0.25)
# metode ini mebagi data test dan data training secara random, sehingga hasil dapat berubah-ubah, hal ini diatasi dengan parameter random_state, nilainya boleh berapa saja berupa integer

x_train, x_test, y_train, y_test = tts(liter, jarak, random_state=21, test_size=0.2)
```

```python
plt.scatter(x_train, y_train)
plt.xlabel("Bahan Bakar (liter)")
plt.ylabel("Jarak Tempuh (kilometer)")
plt.title("Konsumsi Bahan Bakar")
plt.show()
```

```python
model_l = lr()

#fit digunakan untuk melatih model
model_l.fit(x_train, y_train)
```

```python
#memprediksi hasil model
y_pred = model_l.predict(x_test)
```

```python
#evaluasi model menggunakan R^2, prediksi y_test yang sempurna menghasilkan nilai R^2 = 1.
r2 = metrics.r2_score(y_test, y_pred)
print("The model perfomance for testing set")
print("------------------------------------")
print("R2 score is {}".format(r2))
```
